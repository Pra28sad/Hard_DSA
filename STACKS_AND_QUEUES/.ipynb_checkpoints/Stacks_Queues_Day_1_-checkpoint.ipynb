# 1) Infix to Postfix

Problem Statement: Given an infix expression, Your task is to convert the given infix expression to a postfix expression.

Examples:

Example 1:
Input: a+b*(c^d-e)^(f+g*h)-i
Output: abcd^e-fgh*+^*+i-
Explanation: Infix to postfix

Example 2:
Input: (p+q)*(m-n)
Output: pq+mn-*
Explanation: Infix to postfix
Solution

```python
class Stack:
    def __init__(self, capacity):
        self.top = -1
        self.arr = [None] * capacity
        self.capacity = capacity
        
    def isEmpty(self):
        return self.top == -1
    
    def isFull(self):
        return self.top == self.capacity - 1
    
    def push(self, data):
        if self.isFull():
            print("Stack Overflow")
            return
        self.top += 1
        self.arr[self.top] = data
        
    def pop(self):
        if self.isEmpty():
            print("Stack Underflow")
            return None
        data = self.arr[self.top]
        self.arr[self.top] = None
        self.top -= 1
        return data
    
    def peek(self):
        if self.isEmpty():
            print("Stack is empty")
            return None
        return self.arr[self.top]
    
    def traverse(self):
        if self.isEmpty():
            print("Stack is empty")
            return
        print("Stack elements (top â†’ bottom): ", end="")
        for i in range(self.top, -1, -1):
            print(self.arr[i], end=" ")
        print()


# Example usage
stack = Stack(5)
stack.push(10)
stack.push(20)
stack.push(30)
stack.traverse()      # Output: Stack elements (top â†’ bottom): 30 20 10
print(stack.peek())   # Output: 30
print(stack.pop())    # Output: 30
stack.traverse()      # Output: Stack elements (top â†’ bottom): 20 10
```

```python
class Stack:
    def __init__(self,capacity):
        self.top = -1
        self.arr = [None]*capacity
        self.capacity = capacity
        
    def isEmpty(self):
        return self.top == -1
    
    def isFull(self):
        return self.top == self.capacity - 1
    
    def push(self,data):
        if self.isFull():
            print("Stack overflow")
            return 
        self.top+= 1
        self.arr[self.top] = data
        
    def pop(self):
        if self.isEmpty():
            print("Underflow")
            return
        
        data = self.arr[self.top]
        self.arr[self.top] = None
        self.top -= 1
        
        return data
    
    def peek(self):
        if self.isEmpty():
            return None
        
        return self.arr[self.top]
    
    
def precedence(op):
    if op == "^":
        return 3
    elif op in ("*","/"):
        return 2
    elif op in ("+","-"):
        return 1
    else:
        return 0
    
def infix_to_postfix(expression):
    s = Stack(len(expression))
    output = ""
    
    for ch in expression:
        if ch.isalnum(): # operand
            output += ch
        elif ch=="(":
            s.push(ch)
        elif ch==")":
            while not s.isEmpty() and s.peek() != "(":
                output += s.pop()
                
            s.pop()
        else:
            while (not s.isEmpty() and precedence(s.peek()) >= precedence(ch) and ch != "^"): # handle associativity
                
                output += s.pop()
            s.push(ch)
            
    while not s.isEmpty():
        output += s.pop()
    return output


# ðŸ§  Example Test
expr = "a+b*(c^d-e)^(f+g*h)-i"
print("Infix:  ", expr)
print("Postfix:", infix_to_postfix(expr))
```
