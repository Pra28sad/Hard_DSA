```python
class Stack:
    def __init__(self,capacity):
        self.arr = [None]*capacity
        self.capacity = capacity
        self.top = -1
        
    def isEmpty(self):
        return self.top == -1
    
    def isFull(self):
        return self.top == self.capacity - 1
    
    def push(self,data):
        if self.isFull():
            print("Stack Overflow")
            return
        
        self.top += 1
        self.arr[self.top] = data
    
    def pop(self):
        if self.isEmpty():
            print("Stack is empty")
            return
        
        data = self.arr[self.top]
        self.arr[self.top] = None
        self.top -= 1
        
        return data
    
    def peek(self):
        if self.isEmpty():
            print("stack is empty")
            return
        
        return self.arr[self.top]
    
    def traverse(self):
        if self.isEmpty():
            print("Stack is empty")
            return
        print("Stack elements (top → bottom): ", end="")
        for i in range(self.top,-1,-1):
            print(self.arr[i],end=" ")
        print()
        
        
# Example usage
stack = Stack(5)
stack.push(10)
stack.push(20)
stack.push(30)
stack.traverse()      # Output: Stack elements (top → bottom): 30 20 10
print(stack.peek())   # Output: 30
print(stack.pop())    # Output: 30
stack.traverse()      # Output: Stack elements (top → bottom): 20 10
```

# 1) Check for Balanced Parentheses


Problem Statement: Check Balanced Parentheses. Given string str containing just the characters '(', ')', '{', '}', '[' and ']', check if the input string is valid and return true if the string is balanced otherwise return false.

Note: string str is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Example 1:

Input: str = “( )[ { } ( ) ]”

Output: True

Explanation: As every open bracket has its corresponding 
close bracket. Match parentheses are in correct order 
hence they are balanced.
Example 2:

Input: str = “[ ( )”

Output: False

Explanation: As ‘[‘ does not have ‘]’ hence it is 
not valid and will return false.

```python
def check_balanced_paranthesis(expression):
    st = Stack(len(expression))
    pairs = {')': '(', ']': '[', '}': '{'}
    
    for it in expression:
        if it in "([{":
            st.push(it)
        elif it in ")]}":
            if st.isEmpty() or st.peek() != pairs[it]:
                return False
            st.pop()
    
    return st.isEmpty()
```

```python
def isValid(s: str) -> bool:
    st = []
    for it in s:
        if it == '(' or it == '{' or it == '[':
            st.append(it)
        else:
            if len(st) == 0:
                return False
            ch = st[-1]
            st.pop()
            if (it == ')' and ch == '(') or (it == ']' and ch == '[') or (it == '}' and ch == '{'):
                continue
            else:
                return False
    return len(st) == 0
```

# 2) Min Stack

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the MinStack class:

MinStack() initializes the stack object.
void push(int val) pushes the element val onto the stack.
void pop() removes the element on the top of the stack.
int top() gets the top element of the stack.
int getMin() retrieves the minimum element in the stack.
You must implement a solution with O(1) time complexity for each function.

 

Example 1:

Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
 

Constraints:

-231 <= val <= 231 - 1
Methods pop, top and getMin operations will always be called on non-empty stacks.
At most 3 * 104 calls will be made to push, pop, top, and getMin.

```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []
        
    def push(self,val):
        self.stack.append(val)
        
        if not self.min_stack or val<=self.min_stack[-1]:
            self.min_stack.append(val)
            
        else:
            self.min_stack.append(self.min_stack[-1])
            
    def pop(self):
        if self.stack:
            self.stack.pop()
            self.min_stack.pop()
            
    def top(self):
        return self.stack[-1]
    
    def getMin(self):
        return self.min_stack[-1]
    
    
```

```python
class Node:
    def __init__(self, val, min_val, next=None):
        self.val = val
        self.min_val = min_val
        self.next = next


class MinStack:
    def __init__(self):
        self.head = None  # top of the stack

    def push(self, val):
        """
        :type val: int
        :rtype: None
        """
        if not self.head:
            # Stack empty: min is the val itself
            self.head = Node(val, val)
        else:
            # Compute new min
            current_min = min(val, self.head.min_val)
            # Create new node and make it new head
            new_node = Node(val, current_min, self.head)
            self.head = new_node

    def pop(self):
        """
        :rtype: None
        """
        if self.head:
            self.head = self.head.next

    def top(self):
        """
        :rtype: int
        """
        if self.head:
            return self.head.val

    def getMin(self):
        """
        :rtype: int
        """
        if self.head:
            return self.head.min_val
```

```python
class MinStack:
    def __init__(self):
        self.stack = []      # stores all values
        self.min_stack = []  # stores only current minimums

    def push(self, val: int) -> None:
        self.stack.append(val)
        # Only push to min_stack if it's a new minimum
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self) -> None:
        if not self.stack:
            return
        val = self.stack.pop()
        # If the popped element was the current minimum, pop it from min_stack too
        if val == self.min_stack[-1]:
            self.min_stack.pop()

    def top(self) -> int:
        if self.stack:
            return self.stack[-1]

    def getMin(self) -> int:
        if self.min_stack:
            return self.min_stack[-1]
```
