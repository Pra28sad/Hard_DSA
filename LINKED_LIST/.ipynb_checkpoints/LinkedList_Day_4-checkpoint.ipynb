# 1) Add Two Numbers

Problem Statement: Given the heads of two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

Examples:

Input Format: 
(Pointer/Access to the head of the two linked lists)

num1  = 243, num2 = 564

l1 = [2,4,3]
l2 = [5,6,4]

Result: sum = 807; L = [7,0,8]

Explanation: Since the digits are stored in reverse order, reverse the numbers first to get the or                                                original number and then add them as â†’ 342 + 465 = 807. Refer to the image below.

```python
class Node:
    def __init__(self,data,Next=None):
        self.data = data
        self.next = Next
        
    def setNext(self,Next):
        self.next = Next
        
    def getNext(self):
        return self.next
    
    def setData(self,data):
        self.data = data
        
    def getData(self):
        return self.data
    
    
```

```python
class Node:
    def __init__(self,data,Next=None):
        self.data = data
        self.next = Next
        
def addTwoNumbers(l1,l2):
    dummy = Node(0)
    curr = dummy
    carry  = 0
    
    while l1 or l2 or carry:
        val1 = l1.data if l1 else 0
        val2 = l2.data if l2 else 0
        print(val1,val2)
        
        total = val1+val2+carry
        
        carry = total//10
        
        curr.next = Node(total%10)
        curr = curr.next
        
        if l1:
            l1 = l1.next
            
        if l2:
            l2 = l2.next
            
    return dummy.next

def traverse(head):
    temp = head
    
    while temp:
        print(temp.data, end="->  " if temp.next else "")
        temp=temp.next
        
    print()
    

def build_linked_list(values):
    head = Node(values[0])
    temp = head
    
    for val in values[1:]:
        temp.next = Node(val)
        temp = temp.next
        
    return head

l1 = build_linked_list([2,4,3])
l2 = build_linked_list([5,6,4])

res = addTwoNumbers(l1,l2)
traverse(res)
```
