# Introduction to Doubly Linked List

```python
class DNode:
    def __init__(self,data, prev=None, next=None):
        self.data = data
        self.prev = prev
        self.next = next
        
    def setNext(self,next_node):
        self.next = next_node
        
    def getNext(self):
        return self.next
    
    def setPrev(self,prev_node):
        self.prev = prev_node
        
    def getPrev(self):
        return self.prev
    
    def getData(self):
        return self.data
    
    def setData(self):
        self.data = data
        
        
head = DNode(1)
node2 = DNode(2)
node3 = DNode(3)
node4 =  DNode(4)

#3 linking nodes to forward
head.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)

# linking nodes backward
node2.setPrev(head)
node3.setPrev(node2)
node4.setPrev(node3)

```

```python
## frontward
def travverse(head):
    current = head
    while current:
        print(current.getData(),end=" -> ")
        current = current.getNext()
        
    print("None")
    
travverse(head)
```

```python
## backward

def reTraverse(tail):
    current = tail
    while current:
        print(current.getData(),end="-> ")
        current = current.getPrev()
    print("None")
    
reTraverse(node4)
```

# 2) Reverse the double linkedlist


```python
class DNode:
    def __init__(self, data, prev=None, next=None):
        self.data = data
        self.prev = prev
        self.next = next

    def setNext(self, next_node):
        self.next = next_node

    def getNext(self):
        return self.next

    def setPrev(self, prev_node):
        self.prev = prev_node

    def getPrev(self):
        return self.prev

    def getData(self):
        return self.data

    def setData(self, data):
        self.data = data


def reverse_doubly_linked_list(head):
    """Reverse a doubly linked list and return the new head."""
    current = head
    new_head = None
    
    while current:
        # Swap prev and next
        current.prev, current.next = current.next, current.prev
        
        # Move to the next node (which is current.prev now, because swapped)
        new_head = current
        current = current.prev
    
    return new_head


# Example usage
head = DNode(1)
node2 = DNode(5)
node3 = DNode(7)
node4 = DNode(8)

# Linking forward
head.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)

# Linking backward
node2.setPrev(head)
node3.setPrev(node2)
node4.setPrev(node3)

# Traverse forward before reverse
print("Forward before reverse:")
cur = head
while cur:
    print(cur.getData(), end=" -> ")
    cur = cur.getNext()
print("None")

# Reverse
head = reverse_doubly_linked_list(head)

# Traverse forward after reverse
print("\nForward after reverse:")
cur = head
while cur:
    print(cur.getData(), end=" -> ")
    cur = cur.getNext()
print("None")
```
