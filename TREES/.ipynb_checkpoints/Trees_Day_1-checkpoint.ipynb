# 1) Inorder Traversal of Binary Tree

### Problem Statement: 
Given the root of a Binary Tree, write a recursive function that returns an array containing the inorder traversal of the tree.

 Example 1:
				Input: Binary Tree: 4 2 5 3 -1 7 6 -1 9 -1 -1 8 -1 1
				


Output: Output: [3, 1, 9, 2, 4, 7, 5, 8, 6]
				Explanation: We traverse the binary tree in the order of Left, Root then Right recursively resulting in the following traversal:
                
     

Example 2:
				Input: Binary Tree: 1 2 3 4 5 6 7 -1 -1 8 -1 -1 -1 9 10
                


Output : [1 2 4 5 8 3 6 7 9 10]
				Explanation: We traverse the binary tree in the order of Root, Left then Right recursively resulting in the following traversal:
                


```python
# Node class for the binary tree
class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None

# Function to perform inorder traversal
# of the tree and store values in 'arr'
def inorder(root, arr):
    # If the current node is None
    # (base case for recursion), return
    if root is None:
        return
    # Recursively traverse the left subtree
    inorder(root.left, arr)
    # Append the current node's
    # value into the list
    arr.append(root.data)
    # Recursively traverse 
    # the right subtree
    inorder(root.right, arr)

# Function to initiate inorder traversal
# and return the resulting list
def in_order(root):
    # Create an empty list to
    # store inorder traversal values
    arr = []
    # Call the inorder traversal function
    inorder(root, arr)
    # Return the resulting list
    # containing inorder traversal values
    return arr

# Main function
if __name__ == "__main__":
    # Creating a sample binary tree
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)

    # Getting inorder traversal
    result = in_order(root)

    # Displaying the inorder traversal result
    print("Inorder Traversal:", end=" ")
    # Output each value in the
    # inorder traversal result
    for val in result:
        print(val, end=" ")
    print()
             
```

```python
## leet code >>

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[int]
        """
        if not root:
            return []

        ## recursively get values from left subtree
        left = self.inorderTraversal(root.left)

        ## current  nodee
        current = [root.val]

        ## recursively get values from right subtree
        right = self.inorderTraversal(root.right)

        ## combine results
        return left+current+right

        
```

# 2) Iterative Postorder Traversal of Binary Tree Using 2 Stacks

##### Problem Statement:
Given the root of a Binary Tree, create a function that performs a postorder traversal using two stacks and returns an array containing the traversal sequence.

 Example 1:
				Input: Binary Tree: 4 2 5 3 -1 7 6 -1 9 -1 -1 8 -1 1
				



Output: [1, 9, 3, 2, 7, 8, 6, 5, 4] 
				Explanation: We traverse the binary tree in the order of Left, Right then Root recursively resulting in the following traversal:
                


Example 2:
				Input:Binary Tree: 1 2 3 4 5 6 7 -1 -1 8 -1 -1 -1 9 10



Output : [1 2 4 5 8 3 6 7 9 10]
				Explanation:  We traverse the binary tree in the order of Left, Right then Root recursively resulting in the following traversal:



##### result : 



```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
        
    def setData(self,data):
        self.data =data
    
    def getData(self):
        return self.data
    
    def setNext(self,next):
        self.next = next
        
    def getNext(self):
        return self.next
    
    
class StackLL:
    def __init__(self):
        self.head = None
        
    def isEmpty(self):
        return self.head == None
    
    def push(self,data):
        node = Node(data)
        node.setNext(self.head)
        self.head = node
        
    def pop(self):
        if self.isEmpty():
            print("stack was empty")
            return
        
        data = self.head.getData()
        self.head = self.head.getNext()
        return data
    
    def traverse(self):
        if self.isEmpty():
            print("stack was empty")
            return
        
        temp = self.head
        while temp:
            print(temp.getData(),end="--<")
            temp = temp.getNext()
            
    def peak(self):
        if self.isEmpty():
            print("Stacl is empty")
            return
        
        data = self.head.getData()
        return data
    
    def size(self):
        if self.isEmpty():
            print("empty")
            return
        
        size = 0
        temp = self.head
        while temp:
            size += 1
            temp = temp.getNext()
            
        return size
    
 
# Node class for the binary tree
class Tree:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None
        



```

```python
## now we will write actuall sum:

def postOrder_traversal(root):
    if root is None:
        return []
    
    ## empty stacks
    stack1 = StackLL()
    stack2 = StackLL()
    
    res = []
    
    #pushing the first root
    stack1.push(root)
    
    #running while stack1 is not empty
    while not stack1.isEmpty():
        node = stack1.pop()
        stack2.push(node)
        
        ## push left and right children of removed item to stack1
        if node.left:
            stack1.push(node.left)
            
        if node.right:
            stack1.push(node.right)
            
        
    ## pop all items from stack2 and add to result
    while not stack2.isEmpty():
        node = stack2.pop()
        res.append(node.data)
        
    return res

# Build the binary tree
root = Tree(4)
root.left = Tree(2)
root.right = Tree(5)
root.left.left = Tree(3)
root.left.right = Tree(7)
root.right.left = Tree(6)
root.left.left.right = Tree(9)
root.right.left.left = Tree(8)
root.right.left.right = Tree(1)

# Perform traversal
print("Postorder Traversal:", postOrder_traversal(root))
```

# 3) Post - Order Traversal of binary tree

Problem Statement: Given the root of a Binary Tree, write a recursive function that returns an array containing the postorder traversal of the tree.

Examples
Input: Binary Tree: 4 2 5 3 -1 7 6 -1 9 -1 -1 8 -1 1

Output: [1, 9, 3, 2, 7, 8, 6, 5, 4]
Explanation: We traverse the binary tree in the order of Left, Right then Root recursively resulting in the following traversal:

Input: Binary Tree: 1 2 3 4 5 6 7 -1 -1 8 -1 -1 -1 9 10

Output: [1 2 4 5 8 3 6 7 9 10] 
Explanation: We traverse the binary tree in the order of Left, Right then Root recursively resulting in the following traversal:


```python
def postOrder_oneStack(root):
    if root is None:
        return []
    
    stack = StackLL()
    res = []
    last_visited = None
    current = root
    
    while not stack.isEmpty() or current is not None:
        ## go to left most node
        if current:
            stack.push(current)
            current = current.left
            
        else:
            peek = stack.peak()
            
            ## if right child exists and not yet visited -- >move right
            if peek.right and last_visited != peek.right:
                current = peek.right
                
            else:
                
                ## visit the node
                node = stack.pop()
                res.append(node.data)
                last_visited = node
                
    return res
# Build a binary tree manually
root = Tree(4)
root.left = Tree(2)
root.right = Tree(5)
root.left.left = Tree(3)
root.left.right = Tree(7)
root.right.left = Tree(6)
root.left.left.right = Tree(9)
root.right.left.left = Tree(8)
root.right.left.right = Tree(1)

# Perform traversal
print("Postorder Traversal (1 StackLL):", postOrder_oneStack(root))

```

# 4) PreOrder InOrder PostOrder Traversals in One Traversal

##### Problem Statement:
    Given the root of a Binary Tree, return the preorder, inorder and postorder traversal sequence of the given tree by making just one traversal
  


```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next
        
    def setData(self, data):
        self.data = data
    
    def getData(self):
        return self.data
    
    def setNext(self, next):
        self.next = next
        
    def getNext(self):
        return self.next
    

class StackLL:
    def __init__(self):
        self.head = None
        
    def isEmpty(self):
        return self.head is None
    
    def push(self, data):
        node = Node(data)
        node.setNext(self.head)
        self.head = node
        
    def pop(self):
        if self.isEmpty():
            print("Stack is empty")
            return None
        data = self.head.getData()
        self.head = self.head.getNext()
        return data
    
    def peak(self):
        if self.isEmpty():
            return None
        return self.head.getData()
    
    def traverse(self):
        temp = self.head
        while temp:
            print(temp.getData(), end=" -> ")
            temp = temp.getNext()
        print()


# Binary tree node
class Tree:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None


# Function for one-pass traversal
def Preorder_Inorder_Postorder(root):
    if not root:
        return [], [], []
    
    s = StackLL()
    s.push((root, 1))
    
    preorder = []
    inorder = []
    postorder = []
    
    while not s.isEmpty():
        node, state = s.pop()
        
        if state == 1:
            # Preorder
            preorder.append(node.data)
            # Increment state and push back
            s.push((node, 2))
            # Left child (if exists)
            if node.left:
                s.push((node.left, 1))
        
        elif state == 2:
            # Inorder
            inorder.append(node.data)
            # Increment state and push back
            s.push((node, 3))
            # Right child (if exists)
            if node.right:
                s.push((node.right, 1))
        
        else:
            # Postorder
            postorder.append(node.data)
    
    return preorder, inorder, postorder

```

```python
# Constructing the tree
#        1
#       / \
#      2   3
#     / \
#    4   5

root = Tree(1)
root.left = Tree(2)
root.right = Tree(3)
root.left.left = Tree(4)
root.left.right = Tree(5)

pre, ino, post = Preorder_Inorder_Postorder(root)

print("Preorder: ", pre)
print("Inorder:  ", ino)
print("Postorder:", post)
```
