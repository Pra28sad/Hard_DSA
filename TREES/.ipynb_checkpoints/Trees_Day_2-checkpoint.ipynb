# 1) Maximum depth of a Binary Tree

Problem Statement:
Given the root of a Binary Tree, return the height of the tree. The height of the tree is equal to the number of nodes on the longest path from root to a leaf.

Input: Binary Tree: 1 2 5 -1 -1 4 6 5



Output: 4
Explanation: 
In the above example, the height of the binary tree is along the longest path from the root node 1 -> 5 -> 4 -> 5. 
				



Input: Binary Tree: 3 1 2




Output: 2



Explanation: 
We traverse the binary tree in the order of Left, RIght then Root recursively resulting in the following traversal:
				


            

```python
def find_depth(root):
    ## level order traversing
    if not root:
        return 0
    
    queue = [root]
    result = []
    
    while queue:
        level_size = len(queue)
        level = []
        
        for i in range(level_size):
            node = queue.pop(0)
            
            level.append(node.val)
            
            if node.left:
                queue.append(node.left)
                
            if node.right:
                queue.append(node.right)
                
        result.append(level)
        
    return len(result)

    
        
        
find_depth(root)
    
    
```

```python
## using recursion 

def max_depth(root):
    if not root:
        return 0
    
    left_depth = max_depth(root.left)
    right_depth =max_depth(root.right)
    
    return 1+max(left_depth,right_depth)

max_depth(root)
```

```python
## another way

def find_depth(root):
    # Empty tree
    if not root:
        return 0
    
    queue = [root]
    depth = 0
    
    while queue:
        depth += 1
        level_size = len(queue)
        
        for _ in range(level_size):
            node = queue.pop(0)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
                
    return depth
```

# 2) Bottom View of a Binary Tree

Problem Statement: Given a Binary Tree, return its Bottom View. The Bottom View of a Binary Tree is the set of nodes visible when we see the tree from the bottom.

 Example 1:
				Input:Binary Tree: 1 2 3 4 10 9 11 -1 5 -1 -1 -1 -1 -1 -1 -1 6
                
   
   
 Output:Bottom View Traversal: [4, 5, 6, 3, 11]
				Explanation: The bottom view of the binary tree would comprise of the nodes that are the last encountered nodes for each vertical index.
		
   
   
   Example 2:
				Input:Binary Tree: 2 7 5 2 6 -1 9 -1 -1 5 11 4 -1
				
                
 
 
 Output : Bottom View: [2 5 6 11 4 9]
				Explanation: 
				



###### expllanation answer approach>> 





```python
## noode class to represent the binary tree

class Node:
    def __init__(self,val):
        self.data = val
        self.left = None
        self.right = None
        
class Solution:
    def bottomView(self,root):
        
        ## result list
        ans=[]
        
        
        if root is None:
            return ans
        
        
        
        ## Dictionary for vertical positions
        mpp = {} ## key = horizental distance, val = node.data
        
        ## queue for bfs using simple pythohn list
        # each element (node,horizental_distance )
        
        queue = [(root,0)]
        
        while queue:
            node, line = queue.pop(0) ### pop front -- > BFS
            
            ## keep upadating to ensure "bottom-most" stays
            mpp[line] = node.data
            
            ## left child :: horizental distance -1
            if node.left:
                queue.append((node.left,line-1))
                
                
            ## right child: horizental distance + 1
            if node.right:
                queue.append((node.right,line+1))
                
        
        ## collect bottom view from sorted order of horizenatal distances
        
        for key in sorted(mpp.keys()):
            ans.append(mpp[key])
            
        return ans
    
    
                
```

```python
# Creating a sample binary tree
root = Node(1)
root.left = Node(2)
root.left.left = Node(4)
root.left.right = Node(10)
root.left.left.right = Node(5)
root.left.left.right.right = Node(6)
root.right = Node(3)
root.right.right = Node(10)
root.right.left = Node(9)

# Creating a Solution object
solution = Solution()

# Get the Bottom View
bottomView = solution.bottomView(root)

print("Bottom View Traversal:")
for x in bottomView:
    print(x, end=" ")
```

# 3) Right / Left View of Binary Tree

##### Problem Statement: 

Given a Binary Tree, return its right and left views.

The Right View of a Binary Tree is a list of nodes that can be seen when the tree is viewed from the right side. The Left View of a Binary Tree is a list of nodes that can be seen when the tree is viewed from the left side.

 Example 1:
                Input:Binary Tree: 1 2 3 4 10 9 11 -1 5 -1 -1 -1 -1 -1 -1 -1 6 


 Output: Left View: [1, 2, 4, 5, 6] , Right View: [1, 3, 11, 5, 6]
                
Explanation:



Example 2:
                Input:Binary Tree: 2 7 5 2 6 -1 9 -1 -1 5 11 4 -1
                


Output :Left View Traversal:[2,7,2,5], Right View Traversal:[2,5,9,4] 
                
Explanation:




```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# --------------------------------------------------------
# LEFT VIEW  Last node of each level when traversing LEFTRIGHT
# --------------------------------------------------------
def left_view(root):
    if not root:
        return []
    
    # Queue for BFS: stores (node, level)
    queue = [(root, 0)]
    
    # Dictionary to store the LAST node encountered at each level
    mpp = {}
    
    while queue:
        node, lev = queue.pop(0)   # Take next node in BFS
        
        # Overwrite: this ensures the LAST node at this level is stored
        mpp[lev] = node.data
        
        # For left view: process LEFT child first
        if node.left:
            queue.append((node.left, lev + 1))
        
        # Then process RIGHT child
        if node.right:
            queue.append((node.right, lev + 1))
    
    # Return nodes in sorted order of levels: level 0, level 1, level 2, ...
    return [mpp[key] for key in sorted(mpp.keys())]


# --------------------------------------------------------
# RIGHT VIEW  Last node of each level when traversing RIGHTLEFT
# --------------------------------------------------------
def right_view(root):
    if not root:
        return []
    
    # Queue for BFS
    queue = [(root, 0)]
    
    # Dictionary to store the LAST node encountered at each level
    mpp = {}
    
    while queue:
        node, lev = queue.pop(0)

        # Keep overwriting so the last node at every level is stored
        if lev not in mpp:
            mpp[lev] = node.data
        
        # For right view: process RIGHT child first
        if node.right:
            queue.append((node.right, lev + 1))
        
        # Then LEFT child
        if node.left:
            queue.append((node.left, lev + 1))
    
    # Return nodes in increasing order of level
    return [mpp[key] for key in sorted(mpp.keys())]


# --------------------------------------------------------
# Combined Left + Right View
# --------------------------------------------------------
def leftRight_view(root):
    return left_view(root), right_view(root)
```

# 4) Check for Symmetrical Binary Tree

Problem Statement: Given a Binary Tree, determine whether the given tree is symmetric or not. A Binary Tree would be Symmetric, when its mirror image is exactly the same as the original tree. If we were to draw a vertical line through the centre of the tree, the nodes on the left and right side would be mirror images of each other.

Example 1:
Input: Binary Tree: 1 2 2 3 4 4 3




Output:True
Explanation: If we were to draw a vertical line through the centre of the tree, dividing it into left and right parts, we observe that the nodes on the left and right sides are mirror images of each other.





Example 2:
Input:Binary Tree: 1 2 2 -1 3 -1 3




Output : False
Explanation: While the tree has a symmetric structure at the first level with nodes 2 and 2, the subtrees under nodes 2 and 2 are not symmetric. If we were to draw a vertical line through the centre of the tree, dividing it into left and right parts, we observe that the nodes on the left and right sides are not mirror images of each other.




            

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        


def isMirror(left,right):
    if not left and not right:  # If both nodes are None  symmetric
        return True
    
    if not left  or not right:   # If only one is None  NOT symmetric
        return False
    
    # Values must match AND children must be mirror of each other
    return (left.data == right.data 
            and
            isMirror(left.left, right.right)
            and
            isMirror(left.right, right.left))

def isSymmetric(root):
    if not root:
        return True
    
    return isMirror(root.left,root.right)

    


    
root = Node(1)
root.left = Node(2)
root.right = Node(2)
root.left.left = Node(3)
root.left.right = Node(4)
root.right.left = Node(4)
root.right.right = Node(3)

isSymmetric(root)
```

# 5) Check if the Binary Tree is Balanced Binary Tree



##### Problem Statement: 
Given a Binary Tree, return true if it is a Balanced Binary Tree else return false. A Binary Tree is balanced if, for all nodes in the tree, the difference between left and right subtree height is not more than 1.

 Example 1:
                Input:Binary Tree: 3 9 20 -1 -1 15 7
                


                Output: True, This is a Balanced Binary Tree.
                Explanation: The difference in the height of left and right subtree is 1 hence the tree is balanced.
                


                Example 2:
                Input:Binary Tree: 1 3 2 5 4 -1 -1 7 6 
                
            


                Output : False, this is not a Balanced Binary Tree.
                Explanation: The difference in the height of left and right subtree is 2 hence the tree is not balanced.
                





```python
def find_depth(root):
    ## level order traversing
    if not root:
        return 0
    
    queue = [root]
    result = []
    while queue:
        level_size = len(queue)
        level = []
        
        for i in range(level_size):
            node = queue.pop(0)
            level.append(node.val)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
                
        result.append(level)
        
    return len(result)


def check_balance_binary_tree(root):
    if not root:
        return True   # FIXED
    
    left_len = find_depth(root.left)
    right_len = find_depth(root.right)
    
    if abs(left_len - right_len) > 1:
        return False
    
    return check_balance_binary_tree(root.left) and check_balance_binary_tree(root.right)
```
